# 외부 데이터 읽고 쓰기

```python
# data 폴더에 있는iris.csv파일을 불러오자.
df = pd.read_csv("data/Iris.csv")
```

```python
df
```

```python
df.info()
```

```python
# Species column을 숫자로 바꿔보자.
set(df['Species']) # 종류 확인
```

```python
# iris-setosa = 0
# iris-versicolor = 1
# iris-virginica = 2
df.loc[df["Species"] == 'Iris-setosa', "Species"] = 0
df.loc[df["Species"] == 'Iris-versicolor', "Species"] = 1
df.loc[df["Species"] == 'Iris-virginica', "Species"] = 2
```

```python
df
```

```python
# 다시 한번 확인
df.info()
```

```python
set(df["Species"])
```

```python
df.to_csv("data/Iris_edited.csv") # csv 파일로 저장
```
