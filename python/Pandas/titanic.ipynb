```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
sex_dummy=pd.get_dummies(df['Sex'])
sex_dummy
```

```python
Embarked=pd.get_dummies(df['Embarked'],prefix='town')
Embarked
```

```python
train=pd.concat([df, Embarked],axis=1)
train.head()
```

```python
train.drop(['Sex', 'Embarked'], axis=1, inplace=True)
train.head()
```

```python
train
```

```python
train['Age'].plot(kind="box")
```

```python
train.describe()
```

```python
train[["Age", "Survived"]].corr()
```

```python
train=pd.concat([train, sex_dummy],axis=1)
train.head()
```

```python
train.drop(['male'], axis=1, inplace=True)
```

```python
train
```

```python
train[["female", "Survived"]].corr()
```
