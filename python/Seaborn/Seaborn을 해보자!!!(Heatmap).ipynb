```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 기본 셋팅
sns.set_style("whitegrid") # 흰배경
penguins = sns.load_dataset("penguins") #iris와 맞먹는 펭귄데이터
penguins
```

# Heatmap
- 정사각형 그림에 데이터에 대한 정도 차이를 색 차이로 보여주는 plot
- 말 그대로 히트맵이기 때문에 열화상카메라로 사물을 찍은 것처럼 정보의 차이를 보여준다.
- pairplot과 비슷하게 feature간 관계를 시각화할 때 많이 사용한다.

```python
# 각 feature간 상관관계를 파악하기 위해 Correlation matrix(상관계수)를 만든다.
penguins.corr() # 숫자만 계산해줌.
```

```python
# heatmap 출력
sns.heatmap(penguins.corr())
```
